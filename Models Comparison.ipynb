```python
import numpy as np
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import os
from config import FEATURES, FORMAT


RESULT_FOLDER = "lidc_data/result/"
models = ['SVC', 'LOGR', 'CART', 'RF', 'MLP']

acc = "_acc"
rec = "_rec"

def plt_box(col, label, name):
    fig = plt.figure()
#     fig = plt.figure(figsize=(12,8))
    fig.suptitle(name)
    ax = fig.add_subplot(111)
    plt.boxplot(col)
    ax.set_xticklabels(label)
    plt.savefig('lidc_data/box_plot/ '+ name + '.svg', format='svg', dpi=1200)
    plt.show()
```

```python
compare = pd.DataFrame(index=models)

dirs = os.listdir(RESULT_FOLDER)

result = []
# This would print all the files and directories
for file in dirs:
    if file.endswith('npy'):
        result = np.load(RESULT_FOLDER + file, allow_pickle=True)

        s = []
        for i in result:
            for key, value in i.items():  # Loop through all dictionary elements in the list
                s.append(value)
        compare[file[:-4]] = s
        
compare
```

```python
mean_all = compare.applymap(lambda x: x.mean())
max_model = mean_all.max()
mean_all.style.highlight_max(color='#add6ff').highlight_max(color='darkorange',axis=None)
```
